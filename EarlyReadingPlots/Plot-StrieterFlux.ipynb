```python
# First we are going to observe how the pump flux changes in relation to changes in region specific species concentrations. 
# Start by defining the variables we will use, and the subsequent equations.
# An issue we run into is having multiple concentration variables all changing, to combat this we can set some variables ...
# ... and vary individual ones on seperate plots.

# We will start with observing how the intracellular concentration of sodium effects the pump flux. Which is ni_0.
# This is the most important, as sodium has the largest population in these cells. And what's significant about this model...
# ... is that sodium cannot pass from the apical region to another region.

import numpy as np
import matplotlib.pyplot as plt

K_na = 0.02692677070828331 # 0.2 * ((1 + ni_1)/8.33)
P = 1.0 # Non-dimensional I believe, set as 1, we are observing shape not values.
na_1 = 0.005 # M
K_k = 0.006054054054054055 # 0.1 * ((1 + na_0)/18.5)
ni_1 = 0.1215 # M
na_0 = 0.120 # M

def SFlux(ni_0, K_na, P, na_1, K_k, ni_1, na_0):
    SFlux = (P * ((ni_0 / (K_na + ni_0))**3) * (na_1 / (K_k + na_1))**2)

    return SFlux
```

```python
# Now we can define the array for the intracellular concentration of sodium ions.

ni_0_values = np.linspace(0.0001, 1, 1000)

SFlux_values = np.zeros((len(ni_0_values), 1))

for i in range(len(ni_0_values)):
    SFlux_values[i] = SFlux(ni_0_values[i], K_na, P, na_1, K_k, ni_1, na_0)

plt.figure(figsize=(10, 6))
plt.plot(ni_0_values, SFlux_values, color='orange', linestyle='--')
plt.xlabel('Intracellular sodium concentration (ni_0) [M]')
plt.ylabel('Sodium-Potassium Pump Flux')
plt.title('Sodium-Potassium Pump Flux vs Intracellular Sodium Concentration')
plt.grid(True)
plt.show()
```
